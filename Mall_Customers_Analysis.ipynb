# Mall Customers Dataset Analysis

This notebook performs exploratory data analysis (EDA) on the Mall Customers dataset.  
We will load the data, check its structure, and visualize key features.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Loading and Basic Info

Load the dataset and check the basic information such as column names, data types, and the first few rows.

```python
# Load dataset
df = pd.read_csv(r'C:\Users\praveen\Downloads\Mall customer analysis\Mall_Customers.csv')
```

```python
#show basic info
print("Dataset Info:")
print(df.info())
```

```python
# Display first 5 rows
print("\nFirst 5 rows:")
print(df.head())
```

## Summary Statistics

Get summary statistics of numerical columns to understand distributions, mean, min, max, etc.

```python
# Summary statistics
print("\nSummary statistics:")
print(df.describe())
```

## Missing Values Check

Check if there are any missing values in the dataset that need handling.

```python
# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())
```

## Data Visualization

Visualize the distribution of key variables and relationships between them.

```python
# Age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
print(df.columns)
```

```python
# Annual Income distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Annual_Income_(k$)'], bins=20, kde=True)
plt.title('Annual Income Distribution (k$)')
plt.show()
```

```python
# Spending Score distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Spending_Score'], bins=20, kde=True)
plt.title('Spending Score Distribution')
plt.show()
```

```python
# Scatter plot of Annual Income vs Spending Score by Gender
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Annual_Income_(k$)', y='Spending_Score', hue='Genre')
plt.title('Annual Income vs Spending Score by Gender')
plt.show()
```

## Conclusion

The initial exploratory analysis shows the distribution of customer age, income, and spending scores.  
Further analysis could include clustering customers into segments or predictive modeling.
